```python
def getcolumnindex(df,columnkey):
    columnindex=0
    for i in range(len(df.columns)):
        if df.columns[i]==columnkey:
            columnindex=i
    
    return columnindex
#getcolumnindex(incomedf,"2007")
#print(incomedf.columns)
```

```python
#3.function to print State with lowest average income from 2007 to 2010(inclusive)
def lowestavgincomebetween2years(a,b):# used also to print highest avg between two years
    sum = 0
    a=getcolumnindex(incomedf,str(a))
    b=getcolumnindex(incomedf,str(b))
    c=0
    d={}
    for i in range(0,5):
        for j in range(a,b+1):
            sum += incomedf.values[i][j]
            c+=1
        #print(c)
        f=sum/c
        g=incomedf.values[i][1]
        d[g]=f
    #print(d)
    for i in d:
        if d[i]==min(d.values()):
            print(i,":",d[i])
            #return i
lowestavgincomebetween2years(2007,2010)
    
    




#incomedf
```
