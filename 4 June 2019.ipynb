```python
#linear graph
import matplotlib.pyplot as plot
y=[1,2,3,4,5,6,7,8,9,0]
x=[5,3,5,0,5,2,5,0,5,0]
plot.title("basic linear graph")
plot.plot(x,y)
plot.show()
```

```python
#graph to plot distance& time
distance=list(range(2,11,2))
time = list(range(1,6))
#print(distance,time)
plot.xlabel("time")
plot.ylabel("distance")
plot.plot(distance,time,"m*")
plot.grid()
plot.show()
```

```python
t=list(range(1,11))
a=list(range(len(t)+1))
v=[a[i]*i for i in t]

plot.plot(t,v,"m*")
plot.grid()
plot.show()
```

```python
n=list(range(10))
sqr=[i**2 for i in n]
plot.plot(n,sqr,"m*")
```

```python
#voltage- time
import numpy as np

t=[i/10 for i in range(1,21)]
v=[1+np.sin(2*np.pi*i) for i in t]
plot.plot(t,v,"m")
plot.grid()
plot.show()
#print(np.pi)
plot.savefig("images/sinegraph.png")
#plot.show()
```

```python
#price income of arkanas from 2005 to 2013
import pandas as pd
df=pd.read_csv("datafiles/income.csv")
#print(df)
years=df.columns[2:]
income=df.values[3][2:]
plot.xlabel("years")
plot.ylabel("income")
plot.plot(years,income)

```
